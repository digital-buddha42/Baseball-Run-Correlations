# Analysis

#### Analysis of various sabermetrics revealed several key insights related to the rate at which runs are scored in the MLB. We looked at offensive stats for a period of 100 years, from 1920-2019, to uncover what facets of offensive play have the highest correlation to runs scored per game. 
##### First, we determined that stealing bases has almost no correlation to runs scored. While the correlation coefficient we calculated reveals this fact, the scatter plot and linear regression we plotted truly illustrate it. 
##### We were also able to determine that batting average and OPS (on-base percentage plus slugging percentage) are both highly positively correlated to runs scored. Given that batting average is calculated as hits divided by total at-bats, and OPS is calculated as on-base percentage (box score divided by total plate appearances) plus slugging percentage (weighted hits divided by total at-bats), we would expect significant positive correlation to runs scored. The results of our analysis indicate that our linear regression equations could be used to approximately predict runs per game. 
##### Finally, we uncovered several trends within our data that inspire further future research and analysis. We can observe changes in our data, sometimes abrupt and dramatic, that appear to have little relationship to any of the stats we analyzed, indicating that other factors may be affecting how runs are scored. Our line plots and box and whisker plots visualize such changes as the continued drop in average runs scored up to the 1960s and the gradual rise after, the sharp decline in stolen bases in the 2000s and 2010s, and the myriad outlier years that appear when looking at average runs and OPS by decade. 





```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
filename = Path("baseball_project_data.csv")
```

# Data Cleaning 

```python
baseball_hitting_df = pd.read_csv(filename)
baseball_hitting_df.head()
```

```python
baseball_hitting_df.columns
```

```python
# Renaming columns

baseball_hitting_df.rename(columns={
    'Year': 'year',
    'R/G': 'runs_per_game',
    'G': 'number_of_games',
    'R': 'runs',
    'H': 'hits',
    'HR': 'homeruns',
    'SB': 'stolen_bases',
    'CS': 'caught_stealing',
    'BB': 'walks',
    'SO':'strikeouts',
    'BA': 'batting_average',
    'OBP': 'on_base_percentage',
    'SLG': 'slugging_percentage',
    'OPS': 'on_base_plus_slugging_percentage',
    'TB': 'total_bases',
    'SH': 'sacrifice_hits',
    'BIP': 'balls_in_play',
    'Tms': 'teams'                 
                  }, inplace=True)
columns = baseball_hitting_df.columns
for column in columns:
    print (column)
```

```python
# Slim it down!

# Get the years we want
cleaned_hitting_df = baseball_hitting_df.iloc[4:104]

cleaned_hitting_df
```

```python
cleaned_hitting_df = cleaned_hitting_df[['year','teams','runs_per_game','number_of_games','runs','hits',
                                         'homeruns','stolen_bases','caught_stealing','walks', 'strikeouts','batting_average',
                                        'on_base_percentage','slugging_percentage','on_base_plus_slugging_percentage',
                                        'total_bases','sacrifice_hits','balls_in_play']]
cleaned_hitting_df.head()
```

```python

# Create a copy of the DataFrame
cleaned_hitting_df_copy = cleaned_hitting_df.copy()

# Perform the calculations and assign the new columns to the copy
cleaned_hitting_df_copy['hits_per_game'] = cleaned_hitting_df_copy['hits'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['stolen_bases_per_game'] = cleaned_hitting_df_copy['stolen_bases'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['caught_stealing_per_game'] = cleaned_hitting_df_copy['caught_stealing'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['homeruns_per_game'] = cleaned_hitting_df_copy['homeruns'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['total_bases_per_game'] = cleaned_hitting_df_copy['total_bases'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['sacrifice_hits_per_game'] = cleaned_hitting_df_copy['sacrifice_hits'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['balls_in_play_per_game'] = cleaned_hitting_df_copy['balls_in_play'] / cleaned_hitting_df_copy['number_of_games']
# Adding a new runs per game column because the old one only has 2 decimal points
cleaned_hitting_df_copy['runs_per_game_new'] = cleaned_hitting_df_copy['runs'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['walks_per_game'] = cleaned_hitting_df_copy['walks'] / cleaned_hitting_df_copy['number_of_games']
cleaned_hitting_df_copy['strikeouts_per_game'] = cleaned_hitting_df_copy['strikeouts'] / cleaned_hitting_df_copy['number_of_games']

cleaned_hitting_df_copy.head()
```

```python
reduced_hitting_df = cleaned_hitting_df_copy[['year','teams','runs_per_game_new','number_of_games','hits_per_game',
                                         'homeruns_per_game','stolen_bases_per_game','caught_stealing_per_game','walks_per_game', 
                                         'strikeouts_per_game','batting_average','on_base_percentage','slugging_percentage',
                                         'on_base_plus_slugging_percentage','total_bases_per_game','sacrifice_hits_per_game',
                                         'balls_in_play_per_game']]
reduced_hitting_df.head()
```

# Data Exploration

```python
correlation_df = reduced_hitting_df.corr()
threshold = 0.8

correlation_df = correlation_df[((correlation_df > threshold) & (correlation_df != 1.0)) |(correlation_df < -threshold)]

correlation_df
```

```python
untouched_df = reduced_hitting_df.corr()
untouched_df
```

## Stolen Bases vs. Runs Per Game (1920-2019)

```python
plt.scatter(reduced_hitting_df["stolen_bases_per_game"], reduced_hitting_df["runs_per_game_new"], marker="o")

plt.xlabel("Stolen Bases Per Game")
plt.ylabel("Runs Per Game")
plt.title("Stolen Bases vs. Runs Per Game (1920-2019)")

(slope, intercept, rvalue, pvalue, stderr) = linregress(reduced_hitting_df["stolen_bases_per_game"], reduced_hitting_df["runs_per_game_new"])
regress_values = reduced_hitting_df["stolen_bases_per_game"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(reduced_hitting_df["stolen_bases_per_game"],regress_values,"r-")
plt.annotate(line_eq,(0.6,3.8),fontsize=15,color="red")

print(f"The r-value is: {rvalue}")
plt.show()
```

## Batting Average vs. Runs Per Game (1920-2019)

```python
plt.scatter(reduced_hitting_df["batting_average"], reduced_hitting_df["runs_per_game_new"], marker="o")

plt.xlabel("Batting Average")
plt.ylabel("Runs Per Game")
plt.title("Batting Average vs. Runs Per Game (1920-2019)")

(slope, intercept, rvalue, pvalue, stderr) = linregress(reduced_hitting_df["batting_average"], reduced_hitting_df["runs_per_game_new"])
regress_values = reduced_hitting_df["batting_average"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(reduced_hitting_df["batting_average"],regress_values,"r-")
plt.annotate(line_eq,(0.26,4.0),fontsize=15,color="red")

print(f"The r-value is: {rvalue}")
plt.show()
```

## OPS vs. Runs Per Game (1920-2019)

```python
plt.scatter(reduced_hitting_df["on_base_plus_slugging_percentage"], reduced_hitting_df["runs_per_game_new"], marker="o")

plt.xlabel("OPS")
plt.ylabel("Runs Per Game")
plt.title("OPS vs. Runs Per Game (1920-2019)")

(slope, intercept, rvalue, pvalue, stderr) = linregress(reduced_hitting_df["on_base_plus_slugging_percentage"], reduced_hitting_df["runs_per_game_new"])
regress_values = reduced_hitting_df["on_base_plus_slugging_percentage"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(reduced_hitting_df["on_base_plus_slugging_percentage"],regress_values,"r-")
plt.annotate(line_eq,(0.68,3.6),fontsize=15,color="red")

print(f"The r-value is: {rvalue}")
plt.show()
```

## Created Bins for Decades 

```python
# Create decade bins 
reduced_hitting_df["decade"] = pd.qcut(reduced_hitting_df["year"], 10, labels = ["1920's", "1930's", "1940's", "1950's",
                                                                                "1960's", "1970's", "1980's", "1990's",
                                                                                "2000's", "2010's"])
reduced_hitting_df
```

```python
decade_runs_per_game = reduced_hitting_df.groupby(["decade"])["runs_per_game_new"].mean()
decade_stolen_bases_per_game = reduced_hitting_df.groupby(["decade"])["stolen_bases_per_game"].mean()
decade_batting_average = reduced_hitting_df.groupby(["decade"])["batting_average"].mean()
decade_ops = reduced_hitting_df.groupby(["decade"])["on_base_plus_slugging_percentage"].mean()

decade_df = pd.DataFrame({"Runs Per Game": decade_runs_per_game,
                             "Stolen Bases Per Game": decade_stolen_bases_per_game,
                             "Batting Average": decade_batting_average,
                             "On Base Plus Slugging Percentage": decade_ops})

# We could try to do some sort of graph with this
decade_df
```
